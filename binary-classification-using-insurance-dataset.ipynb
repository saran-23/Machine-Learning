# Predicting that whether the person buys insurance or not based on their 

```python
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
#assigning the dataset in df
df = pd.read_csv("../input/insurance-dataset/insurance_data.csv")
```

```python
print(df)
```

```python
#printing last five values
df.tail()
```

```python
#visualising the data
plt.scatter(df.age,df.bought_insurance,marker='4',color='green') # here i have chosen age and the bought_insurance features 
df.shape
```

```python
from sklearn.model_selection import train_test_split # to split the dataset 80% train, 20% test
```

```python
x_train, x_test, y_train,y_test = train_test_split(df[['age']],df.bought_insurance,test_size=0.1)
```

```python
x_test
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
model.predict(x_test)
```

```python
#accuracy of the model
model.score(x_test,y_test)
```

```python
#probability of the model 
model.predict_proba(x_test) # for below arr '0'  and '1'
```

```python

model.predict([[39]]) # 1-buys insurance 0-won't buy insurance
```
