```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('/content/dropout-ratio-2012-2015.csv')
```

```python
data.head()
```

```python
data.columns

```

```python
#checking null value if null return true
data.isna() 
```

```python
data.isna().any()
```

```python
#Exploring particular column values by its column name[State_UT] 
data_tn = data[data['State_UT']=='Tamil Nadu' ]
data_ap = data[data['State_UT']=='Andhra Pradesh']
```

```python
data_tn
```

```python
data_ap
```

```python
#checking data type
data_ap.info()
```

```python
#Checking missing values ('NR')
data['Primary_Boys'].values
```

```python
#Replacing missing values
data = data.replace('NR', 0)
```

```python
data['Primary_Boys'].values
```

```python
#Converting to float
data['Primary_Boys']= data['Primary_Boys'].apply(lambda x: float(x))
data['Primary_Girls']= data['Primary_Girls'].apply(lambda x: float(x))
data['Upper Primary_Boys']= data['Upper Primary_Boys'].replace('Uppe_r_Primary',0).apply(lambda x: float(x))
data['Upper Primary_Girls']= data['Upper Primary_Girls'].apply(lambda x: float(x))
data['HrSecondary_Boys']= data['HrSecondary_Boys'].apply(lambda x: float(x))
data['HrSecondary_Girls']= data['HrSecondary_Girls'].apply(lambda x: float(x))
data['Secondary _Boys']= data['Secondary _Boys'].apply(lambda x: float(x))
data['Secondary _Girls']= data['Secondary _Girls'].apply(lambda x: float(x))
```

```python
data['Primary_Boys'].values
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize = (45,25))
sns.set_theme(style="darkgrid")
sns.countplot(data['State_UT'],x="State")
```

```python
#primary boys droupout in AP
plt.bar(data_ap['Primary_Boys'],data_ap['year'],color='green')
plt.xlabel("PRIMARY BOYS",color='purple')
plt.ylabel('YEAR',color='purple')
plt.title("Visualising Andra pradesh's students dropout ratio of primary boys from 2012 - 2015",color='purple')
```

```python
#Total Droupout of Primary boys
plt.bar(data['Primary_Boys'],data['year'],color='green')
plt.xlabel("PRIMARY BOYS",color='purple')
plt.ylabel('YEAR',color='purple')
plt.title("Visualising students dropout ratio of primary boys from 2012 - 2015",color='red')
print('dropout ratio of primary boys is decreasing')
```

```python
#Primary girls dropout in ap
plt.bar(data_ap['Primary_Girls'],data_ap['year'],color='green')
plt.xlabel("PRIMARY GIRLS",color='purple')
plt.ylabel('YEAR',color='purple')
plt.title("Visualising Andra pradesh's students dropout ratio of primary girls from 2012 - 2015",color='blue')
```

```python
#Total Primary girls dropout
plt.bar(data['Primary_Girls'],data['year'],color='green')
plt.xlabel("PRIMARY GIRLS",color='purple')
plt.ylabel('YEAR',color='purple')
plt.title("Visualising Total students dropout ratio of primary girls from 2012 - 2015",color='blue')
```
